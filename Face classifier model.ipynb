# Face Classifier Model
Model goal: to identify the manga volume that the face images are from



## Step 1: Image preprocessing
### Part 1: Hand-labelling images & image data augmentation
Hand-labelling 100 manga face from each manga volume and boost the number with image data augmentation





```python
from itertools import product
from PIL import Image
import numpy as np
```

```python
import random
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util
```

#### Image data augmentation 
Through the hand-labelling process, I realized faces from the same character are very similer with a bit of differences in angels and dimentions. Using data augumemtation, I could boost my input images size more efficiently. 

The methods I chose are:

a. Rotation: rotate the original image with a random angel between (-25$^{\circ}$ to 25$^{\circ}$)

b. Add noise

c. Flip Horizontally

Below are the four functions to achieve these tasks:


```python
#function_1 to rotate the original image
def random_rotation(image_array: ndarray):
    
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)
```

```python
#function_2 to add noise to the original image
def random_noise(image_array: ndarray):
    # add random noise to the image
    return sk.util.random_noise(image_array)
```

```python
#function_3 to flip the original image
def horizontal_flip(image_array: ndarray):
    
    return image_array[:, ::-1]
```

```python
# function_4 to loop the image path and ramdon apply the three augmentation functions
import random
import glob

def image_generator(folder_name, num_files_desired=1000):
    # the folder path containing hand-labelled images
    folder_path = 'cropped/'+str(folder_name)
    
    # loop on all files of the folder and build a list of files paths
    images = glob.glob(folder_path+'/*.tif')


    num_generated_files = 0
    while num_generated_files <= num_files_desired:
        # random image from the folder
        image_path = random.choice(images)
        # read image as an two dimensional array of pixels
        img=Image.open(image_path)
        image_to_transform = np.array(img, dtype="uint8")
        available_transformations = {
        'rotate': random_rotation,
        'noise': random_noise,
        'horizontal_flip': horizontal_flip}

        # random num of transformations to apply
        num_transformations_to_apply = random.randint(1, len(available_transformations))

        num_transformations = 0
        transformed_image = None
        while num_transformations <= num_transformations_to_apply:
            # choose a random transformation to apply for a single image
            key = random.choice(list(available_transformations))
            transformed_image = available_transformations[key](image_to_transform)
            new_file_path = '%s/augmented_image_%s.jpg' % (folder_path, num_generated_files)
            sk.io.imsave(new_file_path, transformed_image)
            num_transformations += 1
        num_generated_files+=1
```

```python
#generate images for each manga volume
image_generator('Nekodama',1000)

image_generator('MiraiSan',1000)

image_generator('RisingGirl',1000)
```

### Part 2: Covert all images to 150x150x1 numpy arrays with true labels

```python
from PIL import Image
import numpy as np
import glob
#function to convert all images and their corresponding labels from given folders to numpy arrays with dimention of (150,150)
def load_image(folders):
    data = []
    target = []
    for i in range(len(folders)):
        images=glob.glob('cropped/'+str(folders[i])+'/*.jpg')
        for image in images:
            img = Image.open(image)
            
            img=img.resize((150,150), Image.ANTIALIAS)
            data.append(np.array(img, dtype="uint8" ))
            target.append(i)
    return data,target
```

```python
folders = ['Nekodama','MiraiSan','RisingGirl']
```

```python
data, target = load_image(folders)
```

```python
data = np.asarray(data).reshape(3003,150,150,1)
data.shape
```

```python
target=np.asarray(target).reshape(3003,1)
target.shape
```

### Part 3: train_val_test split

```python
from sklearn.model_selection import train_test_split
X=data
y=target
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.20, random_state=3)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X_train, y_train, test_size=0.33, random_state=3)
```

## Step 2: CNN model for classfication
### Part 1: model setup





```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

CNN_model = Sequential()

CNN_model.add(Conv2D(32, kernel_size=(3, 3), 
                     kernel_initializer = keras.initializers.glorot_normal(seed = 3),
                     activation = 'relu', 
                     input_shape = (150, 150,1)))
CNN_model.add(Conv2D(16, (3, 3), activation='relu'))

CNN_model.add(MaxPooling2D(pool_size=(20, 20)))

CNN_model.add(Dropout(0.25))
CNN_model.add(Flatten())
CNN_model.add(Dense(128, activation='relu'))
CNN_model.add(Dropout(0.5))
CNN_model.add(Dense(3, activation='softmax'))

CNN_model.summary()
```

```python
CNN_model.compile(loss = 'sparse_categorical_crossentropy',
                  optimizer = 'Adam',
                  metrics = ['accuracy'])
```

### Part 2: Model training and validation

```python
history = CNN_model.fit(X_train, y_train,
              batch_size = 128,
              epochs = 50,
              verbose = 1,
             validation_data = (X_val, y_val))
```

### Part 3: Model performance analysis

```python
#function to plot the accuracy and cost of training history
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
def plot_history(history):
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 1)
    plt.plot(history.history['acc'], label='Training Accuracy')
    plt.plot(history.history['val_acc'], label='Validation Accuracy')
    
    
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(2, 1, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')

    
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.xlabel('epoch')
    
    
    plt.tight_layout()
    plt.savefig('acc_loss_fig.jpg')
    plt.show()
    

```

```python
plot_history(history)

```

After first 5 epochs, the cost of the model started to drop dramatically and gradually flattened after the 20 th epochs. For some of the early epochs, the descending direction improved the accuracy for training set, but reduced the accuracy for validation set.

After the 25th epochs the accuracy for both sets reached 90%.
