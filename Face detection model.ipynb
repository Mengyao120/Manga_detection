# Face Detection Model
Model goal: to detect face images and filter out background



## Step 1: Image preprocessing
### Part 1: Hand-labelling images & image data augmentation
Hand-labelling background images and face images and boost the number with image data augmentation




#### Image data augmentation 
Through the hand-labelling process, I realized faces from the same character are very similer with a bit of differences in angels and dimentions. Using data augumemtation, I could boost my input images size more efficiently. 

The methods I chose are:

a. Rotation: rotate the original image with a random angel between (-25$^{\circ}$ to 25$^{\circ}$)

b. Add noise

c. Flip Horizontally

Below are the four functions to achieve these tasks:


```python
from itertools import product
from PIL import Image
import numpy as np
```

```python
import random
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util
```

```python
#function to rotate the original image
def random_rotation(image_array: ndarray):
    
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)
```

```python
#function to add noise to the original image
def random_noise(image_array: ndarray):
    # add random noise to the image
    return sk.util.random_noise(image_array)
```

```python
#function to flip the original image
def horizontal_flip(image_array: ndarray):
    
    return image_array[:, ::-1]
```

```python
# function to loop the image path and ramdon apply the three augmentation functions
import random
import glob

def image_generator(folder_name, num_files_desired=1000):
    # the folder path containing hand-labelled images
    folder_path = 'cropped/'+str(folder_name)
    
    # loop on all files of the folder and build a list of files paths
    images = glob.glob(folder_path+'/*.jpg')


    num_generated_files = 0
    while num_generated_files <= num_files_desired:
        # random image from the folder
        image_path = random.choice(images)
        # read image as an two dimensional array of pixels
        img=Image.open(image_path)
        image_to_transform = np.array(img, dtype="uint8")
        available_transformations = {
        'rotate': random_rotation,
        'noise': random_noise,
        'horizontal_flip': horizontal_flip}

        # random num of transformations to apply
        num_transformations_to_apply = random.randint(1, len(available_transformations))

        num_transformations = 0
        transformed_image = None
        while num_transformations <= num_transformations_to_apply:
            # choose a random transformation to apply for a single image
            key = random.choice(list(available_transformations))
            transformed_image = available_transformations[key](image_to_transform)
            new_file_path = '%s/augmented_image_%s.jpg' % (folder_path, num_generated_files)
            sk.io.imsave(new_file_path, transformed_image)
            num_transformations += 1
        num_generated_files+=1
```

```python
#generate images for dark background and white background
#face images are random select from the hand_labelled images collected for the classifer model
image_generator('Background',1000)

image_generator('contour',1000)
```

### Part 2: Covert all images to 150x150x1 numpy arrays with true labels

```python
from PIL import Image
import numpy as np
#function to convert all images and their corresponding labels from given folders to numpy arrays with dimention of (150,150)

def load_image(folders):
    data = []
    target = []
    for i in range(len(folders)):
        images=glob.glob('cropped/'+str(folders[i])+'/*.jpg')
        for image in images:
            img = Image.open(image)
            
            img=img.resize((150,150), Image.ANTIALIAS)
            data.append(np.array(img, dtype="uint8" ))
            target.append(i)
    return data,target
```

```python
folders = ['Background','face','contour']
```

```python
data, target = load_image(folders)
data = np.asarray(data).reshape(len(data),150,150,1)
target=np.asarray(target).reshape(len(data),1)
```

### Part 3: train_val_test split

```python
from sklearn.model_selection import train_test_split
X=data
y=target
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X_train, y_train, test_size=0.33, random_state=3)
```

## Step 2: CNN model for detection
### Part 1: model setup



```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

model_bg = Sequential()

model_bg.add(Conv2D(32, kernel_size=(3, 3), 
                    kernel_initializer = keras.initializers.glorot_normal(seed = 4),
                    activation = 'relu', 
                    input_shape = (150, 150,1)))
model_bg.add(Conv2D(16, (3, 3), activation='relu'))

model_bg.add(MaxPooling2D(pool_size=(10, 10)))

model_bg.add(Dropout(0.25))
model_bg.add(Flatten())
model_bg.add(Dense(128, activation='relu'))
model_bg.add(Dropout(0.5))
model_bg.add(Dense(3, activation='softmax'))

model_bg.summary()
```

```python
model_bg.compile(loss = 'sparse_categorical_crossentropy',
                  optimizer = 'Adam',
                  metrics = ['accuracy'])
```

### Part 2: Model training and validation

```python

history=model_bg.fit(X_train, y_train,
              batch_size = 128,
              epochs = 20,
              verbose = 1,
              validation_data = (X_val, y_val))
```

### Part 3: Model performance analysis



```python
#function to plot the accuracy and cost of training history
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
def plot_history(history):
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 1)
    plt.plot(history.history['acc'], label='Training Accuracy')
    plt.plot(history.history['val_acc'], label='Validation Accuracy')
    
    
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(2, 1, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')

    
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.xlabel('epoch')
    
    
    plt.tight_layout()
    plt.savefig('acc_loss_fig_1.jpg')
    plt.show()
```

```python
plot_history(history)
```

The first 5 epochs, the cost of the model are about 10 for both train and validation set. For some of the early epochs, the descending direction improved the accuracy for training set, but reduced the accuracy for validation set.

After the 15th epochs the accuracy for both sets reached 90%.

### Part 1: design a function to segment the image into squared ROI using sliding window approach
*ROI: region of interest

```python
#function to crop the image based on the location coordinates and convert it into 150x150 numpy array
def crops(image_obj,coords):
   
    cropped_image = image_obj.crop(coords)
    cropped_image= cropped_image.convert('L')
    cropped_image=cropped_image.resize((150,150), Image.ANTIALIAS)
    cropped_image=np.array(cropped_image, dtype="uint8")
    return cropped_image   
```

```python
#function to segment the images and save the segments as 150x150 numpy arrays
def input_segments(image_path):
    #open image
    image_obj = Image.open(image_path)
    #find the width and height of the image
    W, H = image_obj.size
    input_array=[]
    coords=[]
    #generate the combinations of coordinates for ROI of size (100X100, 150x150,...,400X400)
    for i in range(100,401,50):
        print(i)
        #the step size for each sliding window is 1/5 of the ROI width and 1/2 of the ROI height 
        X=np.arange(1,W-i,i/5)
        Y=np.arange(1,H-i,i/2)
        X_Y=list(product(X, Y))
        
        for n in range(len(X_Y)):
            (a,b)=X_Y[n]
            coords.append((a,b,a+i,b+i))
    #call the crops function to crop and convert the ROI to 150x150 numpy array
    for coord in coords:
        input_array.append(crops(image_obj,coord))
    input_array = np.asarray(input_array).reshape(len(input_array),150,150,1)
    return input_array
```
